### Checking th e percentage of Null values in dataset

```python
((data.isnull().sum()/len(data))*100).sort_values(ascending=False)
```

##### Dropping columns with more than 50% null values

```python
cols = ['RATE_INTEREST_PRIVILEGED','RATE_INTEREST_PRIMARY','AMT_DOWN_PAYMENT','RATE_DOWN_PAYMENT']
data.drop(cols,axis=1,inplace=True)
```

### Handling columns with more than 20% null values

```python
null = data.isnull().sum()/len(data)*100
null[null > 20]
```

```python
data['NAME_TYPE_SUITE'].unique()
```

```python
data['NAME_TYPE_SUITE'].fillna('Unknown',inplace=True)
```

### Handling Negative values

```python
day_cols = []
for col in data.columns:
    if col.startswith('DAYS'):
        day_cols.append(col)
day_cols
```

```python
for col in day_cols:
    data[col] = abs(data[col])
```

### Replacing Null values in Day columns with 0

```python
for col in day_cols:
    data[col] = data[col].replace(np.nan,0)
```

##### Replacing null values in Continuous Numerical columns 

```python
cont_cols = ['AMT_GOODS_PRICE','AMT_ANNUITY','AMT_CREDIT']
for col in cont_cols:
    data[col].fillna(data[col].median(),inplace=True)
```

```python
null_mask = data[data['CNT_PAYMENT'].isnull()]
null_mask['NAME_CONTRACT_STATUS'].value_counts()
```

The corresponding 'CONTRACT STATUS' for null values in the column 'CNT_PAYMENT',shows most of the loans where cancelled,refused or unsused. Hence replacing the null values in CNT_PAYMENT with 0

```python
data['CNT_PAYMENT'].fillna(0,inplace=True)
```

##### Replacing null values in Categorical columns 

```python
data['PRODUCT_COMBINATION'].fillna(data['PRODUCT_COMBINATION'].mode()[0],inplace=True)
```

```python
null_mask1 = data[data['NFLAG_INSURED_ON_APPROVAL'].isnull()]
null_mask1['NAME_CONTRACT_STATUS'].value_counts()
```

Similarly for the null values in 'NFLAG INSURED ON APPROVAL', the corresponging 'NAME_CONTRACT_STATUS' is either of cancelled or refused or unsused. Hence replacing the null values with 'N'

```python
data['NFLAG_INSURED_ON_APPROVAL'].fillna(0.,inplace=True)
```

### Correcting the data types

```python
for col in day_cols:
    data[col]= data[col].astype(int)
data['NFLAG_INSURED_ON_APPROVAL'] = data['NFLAG_INSURED_ON_APPROVAL'].astype(int)
```

#### Dropping the Unnecessary Columns

```python
cols1 = ['WEEKDAY_APPR_PROCESS_START','HOUR_APPR_PROCESS_START','FLAG_LAST_APPL_PER_CONTRACT','NFLAG_LAST_APPL_IN_DAY']
data.drop(cols1,axis=1,inplace=True)
```

## Outliers Detection

```python
data.select_dtypes(exclude=object).columns
```

```python
num_cols = ['AMT_ANNUITY', 'AMT_APPLICATION','AMT_CREDIT', 'AMT_GOODS_PRICE', 'DAYS_DECISION', 'SELLERPLACE_AREA','CNT_PAYMENT','DAYS_FIRST_DRAWING', 'DAYS_FIRST_DUE',
       'DAYS_LAST_DUE_1ST_VERSION', 'DAYS_LAST_DUE', 'DAYS_TERMINATION']
len(num_cols)
```

## Boxplots of Numerical columns

```python
# initialize figure with 10 subplots in 2 rows
fig, axes = plt.subplots(2, 6,figsize=(20,12))
fig.suptitle('Plotting the Boxplot with Outliers for all the numerical columns')
# add padding between the subplots
plt.subplots_adjust(wspace=1) 

# draw boxplot for each column in the each subplot
for i,val in zip(axes.flatten(),num_cols):
        sns.boxplot(data=data[val], ax =i,color = 'purple')       
```

### Outlier Treatment with IQR method

```python
for column in num_cols:
    # IQR Detection
    # Calculate the upper and lower limits
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    data[column] = data[column].clip(lower, upper)
```

### After Outlier Treatment

```python
# initialize figure with 10 subplots in 2 rows
fig, axes = plt.subplots(2, 5,figsize=(15, 12))
fig.suptitle('Plotting the Boxplot with Outliers for all the numerical columns')
# add padding between the subplots
plt.subplots_adjust(wspace=1) 

# draw boxplot for each column in the each subplot
for i,val in zip(axes.flatten(),num_cols):
        sns.boxplot(data=data[val], ax=i,color = 'green')
        
```
